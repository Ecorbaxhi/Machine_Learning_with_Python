```python
# Question 1

# Load the College_new.csv file on Blackboard into a dataframe called college (or give it any name you want)

# Solution

# We can start with importing the necessary library and then load the DataSet. After we can rename the DataSet
# and display the info of this DataSet
import pandas as pd
from google.colab import files
uploaded = files.upload()

college = pd.read_csv("College_new.csv")
college.info()

```

```python
# Question 2

# Check if there are any missing values.

# Solution

# Importing the library
import numpy as np

# Check for missing values
print("Missing values in each column:")
print(college.isnull().sum())
```

```python
# Replace any missing values for numerical columns with the mean value for the column rounded up to the nearest integer

# Solution

# Separate numerical and categorical columns
numerical_cols = college.select_dtypes(include=[np.number]).columns
categorical_cols = college.select_dtypes(include=['object', 'category']).columns
```

```python
# Replace any missing values for categorical columns with the mode of that column.

#Solution

# Fill numerical NaNs with mean (rounded up)
for col in numerical_cols:
    if college[col].isnull().any():
        college[col].fillna(np.ceil(college[col].mean()), inplace=True)

# Fill categorical NaNs with mode
for col in categorical_cols:
    if college[col].isnull().any():
        college[col].fillna(college[col].mode()[0], inplace=True)

# Double-check if any missing values remain
print("\nMissing values after imputation:")
print(college.isnull().sum())
```

```python
# Question 3

# Remove the column that contains college names from the dataframe as it is not useful in prediction.

# Solution

# Let's display first the column names to find the one with college names
print(college.columns)
```

```python
# In this dataset, "College" is an identifier and not a predictor, so we remove it.
college_cleaned = college.drop(columns=['College'])

print(college_cleaned.columns)
```

```python
# Question 4

# Separate the dependent variable and independent variables into two different dataframes

# Solution

# Separate predictors (independent variables) and target (dependent variable)
X = college_cleaned.drop(columns=['Grad.Rate'])
y = college_cleaned['Grad.Rate']

print("Independent Variables (X):")
print(X.head())

print("\nDependent Variable (y):")
print(y.head())
```

```python
# Question 5

# Split the dataset into 80% training and 20% test.

# Solution

# Importing the function to spilt the data
from sklearn.model_selection import train_test_split

# 80% training and 20% testing
X_encoded = pd.get_dummies(X, drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Check the shapes to check that the split happened
print("Training set shape:", X_train.shape, y_train.shape)
print("Test set shape:", X_test.shape, y_test.shape)
```

```python
# Question 6

# Fit a linear multiple regression model to the training set to predict graduation rate using all the other
# features/variables. Use the statsmodels library.

# Solution

# Importing the needed libraries
import statsmodels.api as sm
from sklearn.metrics import mean_absolute_percentage_error

# Encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# To make sure data is clean we convert to numeric and drop NaNs early
X_encoded = X_encoded.apply(pd.to_numeric, errors='coerce')
y_numeric = pd.to_numeric(y, errors='coerce')

X_encoded = X_encoded.dropna()
y_numeric = y_numeric.loc[X_encoded.index]  # match indices

# Split dataset creating training and tests sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_numeric, test_size=0.2, random_state=42)

# Add constant and force all to float
X_train_sm = sm.add_constant(X_train).astype(float)
y_train = y_train.astype(float)

# Fit the model
model = sm.OLS(y_train, X_train_sm).fit()
```

```python
# What variables are significant at the p < 0.05 level?

# Solution

# Print significant variables
significant_vars = model.pvalues[model.pvalues < 0.05].index.tolist()
print("Significant Variables (p < 0.05):", significant_vars)
```

```python
# Predict the graduation rate using the test dataset and report the Mean Absolute Percentage Error (MAPE)
# for the test dataset.

# Solution

# Prepare test data
X_test = X_test.apply(pd.to_numeric, errors='coerce').dropna()
y_test = pd.to_numeric(y_test, errors='coerce').loc[X_test.index]

X_test_sm = sm.add_constant(X_test).astype(float)
y_test = y_test.astype(float)

# Predict and calculate MAPE
y_pred = model.predict(X_test_sm)
mape = mean_absolute_percentage_error(y_test, y_pred)
print("MAPE on Test Set:", round(mape * 100, 2), "%")

# Show summary
model.summary()
```

```python
# Question 7

# Perform backward subset selection to select 7 features using the metric Mean Absolute Percentage Error
# (MAPE) and 3-fold cross validation. Which 7 features are selected? Are the features chosen by the subset
# selection different from the list of statistically significant (p<0.05) variables from fitting the
# linear regression model above? If so, which ones are different?

# Solution

# Importing the missing libraries
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score

# Encode categorical variables and clean
X_encoded = pd.get_dummies(X, drop_first=True)
X_encoded = X_encoded.apply(pd.to_numeric, errors='coerce').dropna()
y_numeric = pd.to_numeric(y, errors='coerce').loc[X_encoded.index]
```

```python
# Perform backward subset selection to select 7 features using the metric Mean Absolute Percentage Error
# (MAPE) and 3-fold cross validation.

# Use SequentialFeatureSelector for backward selection
model = LinearRegression()
selector = SequentialFeatureSelector(
    model,
    n_features_to_select=7,
    direction='backward',
    scoring='neg_mean_absolute_percentage_error',
    cv=3
)

selector.fit(X_encoded, y_numeric)
```

```python
# Which 7 features are selected?

# Extract selected features
selected_features = X_encoded.columns[selector.get_support()].tolist()

print("\n Selected 7 features after backward elimination (sklearn):", selected_features)
```

```python
# Are the features chosen by the subsetselection different from the list of statistically significant
# (p<0.05) variables from fitting the linear regression model above? If so, which ones are different?

# Comparison with significant features from Question 6
significant_from_q6 = ['Apps', 'Top25perc', 'P.Undergrad', 'Outstate', 'Room.Board', 'perc.alumni', 'Expend']

# Check which ones differ
selected_set = set(selected_features)
q6_set = set(significant_from_q6)

print("\nFeatures in Q6 but not in backward selection:", q6_set - selected_set)
print("Features in backward selection but not in Q6:", selected_set - q6_set)
```

```python
# Question 8.

# Create a new dataframe with the 7 features selected above from Qn 7. Fit a linear multiple regression model
# to the training dataset to predict graduation rate using all the other features/variables in the new dataframe.

# Solution

# Import needed library
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error

# Use selected features from Q7 (already selected with SequentialFeatureSelector)
selected_features = X_encoded.columns[selector.get_support()].tolist()

# Create new dataset with selected features
X_selected = X_encoded[selected_features]
y_selected = y_numeric

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_selected, test_size=0.2, random_state=42)
```

```python
# Use the Scikit Learn library this time. Predict the graduation rate using the test dataset
# and report the test Mean Absolute Percentage Error (MAPE) for the test dataset.

# Train model
regr = LinearRegression()
regr.fit(X_train, y_train)

# Predict on test set
y_pred = regr.predict(X_test)

# Evaluate performance (just like your professor's code)
print("R squared:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("MAPE:", round(mean_absolute_percentage_error(y_test, y_pred) * 100, 2), "%")
```

```python
# Question 9

# Fit a ridge regression model on the training dataset for predicting graduation rate using all the predictors,
# with  chosen by 10-fold cross-validation from among the following possibilities for , 0.001, 0.01, 0.05,
# 0.075, 0.1, 1, 5, 10, 100. Predict graduation rate using the test dataset and report the test Mean Absolute
# Percentage Error (MAPE). Report the value of  used in the model.

# Solution

# Importing the missing library
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# We'll use the numeric, encoded predictors and the target.
X_encoded = pd.get_dummies(X, drop_first=True)
# Convert to numeric and drop any rows with missing values
X_encoded = X_encoded.apply(pd.to_numeric, errors='coerce').dropna()
y_numeric = pd.to_numeric(y, errors='coerce').loc[X_encoded.index]

# Split the dataset into training (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y_numeric, test_size=0.2, random_state=42)

```

```python
# RIDGE REGRESSION WITH 10-FOLD CROSS-VALIDATION TO SELECT LAMBDA (alpha)

# Define candidate values for lambda (alpha)
alphas = [0.001, 0.01, 0.05, 0.075, 0.1, 1, 5, 10, 100]

# Set up the Ridge regression estimator and GridSearchCV.
# Note: 'neg_mean_absolute_percentage_error' is used because scikit-learn's
# scoring requires a metric where higher is better. We will negate it to obtain MAPE.
ridge = Ridge()
param_grid = {'alpha': alphas}
grid = GridSearchCV(ridge, param_grid, cv=10, scoring='neg_mean_absolute_percentage_error')
grid.fit(X_train, y_train)
```

```python
# Retrieve the best lambda (alpha)
best_alpha = grid.best_params_['alpha']
print("Best lambda (alpha) chosen:", best_alpha)
```

```python
# PREDICTION AND EVALUATION ON THE TEST SET
# Use the best model to predict the graduation rate on the test set.
y_pred_ridge = grid.predict(X_test)
test_mape = mean_absolute_percentage_error(y_test, y_pred_ridge)
print("Test MAPE for Ridge Regression:", round(test_mape * 100, 2), "%")
```

```python
# PREDICTION AND EVALUATION ON THE TEST SET

# Use the best model to predict the graduation rate on the test set.
y_pred_lasso = grid_lasso.predict(X_test)
test_mape = mean_absolute_percentage_error(y_test, y_pred_lasso)
print("Test MAPE for LASSO Regression:", round(test_mape * 100, 2), "%")
```
